### Veri Ön İşlemeye Giriş

```python
# İlk olarak düşürülecek olarak belirlediğimiz kolonları verisetinden düşürelim.
drop_columns = ['Türü','Kategorisi','Aidat','Kira Getirisi','Site İçerisinde','Fiyat Durumu','Yatırıma Uygunluk','Takas','WC Sayısı','Salon Metrekare','Yapı Tipi',
               'Krediye Uygunluk',]
data_set.drop(labels=drop_columns,axis=1,inplace=True)
```

```python
data_set.head()
```

```python
# Eksik değerleri olan kolonların most_frequence yöntemine göre doldurulması.
from sklearn.impute import SimpleImputer
imp_freq = SimpleImputer(missing_values='Belirtilmemiş',strategy="most_frequent")
```

```python
data_set['Eşya Durumu'] = imp_freq.fit_transform(data_set[['Eşya Durumu']])
```

```python
data_set['Yapı Durumu'] = imp_freq.fit_transform(data_set[['Yapı Durumu']])
```

```python
data_set['Binanın Kat Sayısı'] = imp_freq.fit_transform(data_set[['Binanın Kat Sayısı']])
```

```python
# Net metrekare ve Brüt metrekare özniteliklerinin int tipine dönüştürülmesi
data_set['Net Metrekare'] = data_set['Net Metrekare'].apply(lambda x:int(x.replace('.','').replace(' m2','')))
data_set['Brüt Metrekare'] = data_set['Brüt Metrekare'].apply(lambda x:int(x.replace('.','').replace(' m2','')))
```

```python
# Oda sayısı olarak verilen özniteliğin oda ve salon sayısı olarak ikiye ayrılıp tamsayı tipine dönüştürülmesi
# '8+ Oda' değerine sahip satıların hepsi aynı olduğu için kaldırıyoruz
data_set.drop(labels=data_set.loc[data_set['Oda Sayısı'] == '8+ Oda'].index,axis=0,inplace=True)
data_set.reset_index()
```

```python
data_set['Oda Sayısı'].value_counts()
```

```python
data_set.loc[data_set[data_set['Oda Sayısı']=='1 Oda'].index,'Oda Sayısı'] = '1+0'
```

```python
oda_sayisi = data_set['Oda Sayısı'].apply(lambda x:x.split('+')[0])
salon_sayisi = data_set['Oda Sayısı'].apply(lambda x:x.split('+')[1])

oda_sayisi = oda_sayisi.astype(int)
salon_sayisi = salon_sayisi.astype(int)
```

```python
data_set.insert(4,'Oda Adedi',oda_sayisi)
data_set.drop(labels='Oda Sayısı',axis=1,inplace=True)
```

```python
data_set.insert(5,'Salon Adedi',salon_sayisi)
```

```python
data_set.drop(labels=data_set[data_set.duplicated()].index,axis=0,inplace=True)
```

```python
data_set.reset_index(inplace=True)
```

```python
data_set
```

```python
data_set.drop(labels=['index'],axis=1,inplace=True)
```

```python
data_set.info()
```

```python
data_set['Binanın Kat Sayısı'] = data_set['Binanın Kat Sayısı'].astype(int)
```

```python
data_set.loc[data_set[data_set['Banyo Sayısı']=='6+'].index,'Banyo Sayısı'] = 6
```

```python
data_set['Banyo Sayısı'] = data_set['Banyo Sayısı'].astype(int)
```

```python
data_set.loc[data_set[data_set['Binanın Yaşı'] == '0 (Yeni)'].index,'Binanın Yaşı'] = '0'
```

```python
data_set['Fiyat'] = data_set['Fiyat'].apply(lambda x:x.replace('arrow_downward','').replace('TL','').replace(',',''))
```

```python
data_set['Fiyat'] = data_set['Fiyat'].astype(float)
```

```python
house_data = data_set[['Fiyat','Konum','Bulunduğu Kat','Isıtma Tipi','Binanın Yaşı','Net Metrekare','Brüt Metrekare','Oda Adedi','Binanın Kat Sayısı']]
```

```python
sns.pairplot(house_data)
plt.show()
```

```python
def dummies(x,df):
    temp = pd.get_dummies(df[x], drop_first=True)
    df = pd.concat([df,temp],axis=1)
    df.drop([x],axis=1,inplace=True)
    return df

house_data = dummies('Konum',house_data)
house_data = dummies('Bulunduğu Kat',house_data)
house_data = dummies('Isıtma Tipi',house_data)
house_data = dummies('Binanın Yaşı',house_data)
```

```python
house_data.head()
```

```python
from sklearn.model_selection import train_test_split

np.random.seed(0)
hds_train, hds_test = train_test_split(house_data,train_size=0.8,test_size=0.2,random_state=100)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
num_vars = ['Net Metrekare','Brüt Metrekare','Oda Adedi','Binanın Kat Sayısı','Fiyat']
hds_train[num_vars] = scaler.fit_transform(hds_train[num_vars])
```
